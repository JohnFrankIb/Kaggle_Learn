# Bar Charts and Heatmaps

### Select a Dataset

```python
flight_data = pd.read_csv('flight_delays.csv',index_col='Month')
flight_data.head()
```

### Bar Chart

```python
plt.figure(figsize=(10,6))

sns.barplot(x=flight_data.index,y=flight_data['NK'])

plt.ylabel('Arrival delay (in minutes)')
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')

plt.show()
```

### Heatmap

```python
plt.figure(figsize=(14,7))

sns.heatmap(data=flight_data, annot=True)

plt.xlabel('Airline')
plt.title('Average Arrival Delay for Each Airline, by Month')

plt.show()
```

# Scatter Plots

```python
insurance_data = pd.read_csv('insurance.csv')
insurance_data.head()
```

Simple `Scatter Plot`

```python
sns.scatterplot(x=insurance_data['bmi'],y=insurance_data['charges'])
```

Simple `Regression Plot`

```python
sns.regplot(x=insurance_data['bmi'],y=insurance_data['charges'])
```

### Color-coded scatter plots

```python
sns.scatterplot(x=insurance_data['bmi'],y=insurance_data['charges'],hue=insurance_data['smoker'])
```

`Linear Model Plot`

```python
sns.lmplot(data=insurance_data,x='bmi',y='charges',hue='smoker')
```

```python
sns.swarmplot(data=insurance_data,x='smoker',y='charges')
```

# Distributions

```python
iris_data = pd.read_csv('iris.csv',index_col='Id')
iris_data.head()
```

### `Histograms`

```python
sns.histplot(iris_data['Petal Length (cm)'])
```

### `Density Plots`

Kernel Densiti Estimate (KDE)

```python
sns.kdeplot(data=iris_data['Petal Length (cm)'],shade=True)
```

### `2D KDE Plots`

```python
sns.jointplot(x=iris_data['Petal Length (cm)'],y=iris_data['Sepal Width (cm)'],kind='kde')
```

### `Color-coded Plots`

```python
sns.histplot(data=iris_data,x='Petal Length (cm)',hue='Species')
```

```python
sns.kdeplot(data=iris_data,x='Petal Length (cm)',hue='Species',shade=True)
```
